### 倾向得分计算
使用Logistic回归，计算跨学科的“概率”

```r
# glm拟合，算倾向得分
physicslogit <- glm(
    treat ~ log1p(X10citation) + log1p(coauthor_coun) + log1p(author_avg_age) + log1p(author_avg_coauthor_num)
        + log1p(author_avg_impac) + log1p(author_avg_produc) + log1p(reference_count) + reference_age + JIF,
    data = physicsR, family = "binomial"
)
```

```r
# 多重共线性
car::vif(physicslogit)
```

膨胀系数，全部小于10

```r
# 伪R2
pscl::pR2(physicslogit)["McFadden"]
```

大约能解释12.7%的变异，但伪R2的可靠性不确定。

```r
# 将系数变为odds ratio，大于1表示相较于发生的概率更大
exp(coef(physicslogit))
```

##### 1.2被引量（负二项回归）【伪R2是30.5%】，跨大68%

```r
# 计数变量，且方差远大于均值，所以使用负二项回归；实际上，此处应该有更多的假设需要满足
with(m.dataCitation, tapply(
    X10citation,
    treat, function(x) {
        sprintf("M (SD) = %1.2f (%1.2f)", mean(x), sd(x))
    }
))
```

```r
# 被引量的回归（计数变量，加之方差远大于均值，因此使用nb【负二项回归】）
fitCitation <- glm.nb(X10citation ~ treat + coauthor_coun + author_avg_age + author_avg_coauthor_num +
    log1p(author_avg_impac) + log1p(author_avg_produc) + log1p(reference_count) + reference_age + JIF,
data = m.dataCitation, weights = weights, control = glm.control(maxit = 50)
)
coeftest(fitCitation, vcov. = vcovCL, cluster = ~subclass)
```

```r
# 伪R方
with(summary(fitCitation), 1 - deviance / null.deviance)
```
